# Interactive plot

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

def interactive_plot(amplitude, frequency):
    x = np.linspace(0, 2 * np.pi, 1000)
    y = amplitude * np.sin(frequency * x)
    plt.figure(figsize=(8, 4))
    plt.plot(x, y)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Interactive Sine Wave')
    plt.grid(True)
    plt.show()

interact(interactive_plot, amplitude=(1, 5, 0.1), frequency=(1, 10, 0.1))
```

# Two Frequency Audio and Plot

```python
from ipywidgets import interactive
from IPython.display import Audio, display
```

```python
def beat_freq(f1=220.0, f2=224.0):
    max_time = 3
    rate = 8000
    times = np.linspace(0,max_time,rate*max_time)
    signal = np.sin(2*np.pi*f1*times) + np.sin(2*np.pi*f2*times)
    plt.figure(figsize=(8, 4))
    plt.plot(signal[0:rate])
    plt.xlabel('1 Sec Sample')
    plt.ylabel('Unit Amplitude')
    plt.title("{f1} Hz and {f2} Hz Beats".format(f1=f1, f2=f2))
    display(Audio(data=signal, rate=rate))
    return signal
```

```python
v2 = interactive(beat_freq, f1=(200.0,300.0), f2=(200.0,300.0))
display(v2)
```
